```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Batch_trans = pd.read_csv("./Batch_Parameter_transition.csv")
Epoch_trans = pd.read_csv("./Epoch_Parameter_transition.csv")

Test_data = pd.read_csv("./MoveGenreTest.csv")
Train_data = pd.read_csv("./MoveGenreTrain.csv")
```

```python
Genre_list = list(set(Train_data['Genre']))
Genre_list.sort()
Genre_list.reverse()
print(Genre_list)
```

```python
plt.plot(Batch_trans['Batch_size'], Batch_trans['One_Epoch_Time'])
plt.title("Time change with Batch's size ")
plt.xlabel("Batch_size")
plt.ylabel("Minutes per Epoch ")
plt.show()
```

```python
plt.plot(Epoch_trans['Num_Epoch'], Epoch_trans['Loss_avr'])
plt.xticks([5, 10, 20])
plt.yticks([2 + (i*0.01) for i in range(20)])
plt.xlabel("Num_Epoch")
plt.ylabel("Loss_avr")
plt.title("Loss change with Epoch's transition")
plt.show()
```

```python
countOfGenre = []

for i in range(len(Genre_list)):
    countOfGenre.append(sum(Train_data['Genre'] == Genre_list[i]))
plt.figure(figsize = (10, 10))
plt.barh(Genre_list, countOfGenre)
for i in range(len(Genre_list)):
    plt.annotate(countOfGenre[i], xy = (countOfGenre[i], i), va = "center")
    
plt.ylabel("Genre")
plt.xlabel("Count per Genres")
plt.title("Train Data Count per Genres")
plt.show()
```

```python
countOfGenre = []

for i in range(len(Genre_list)):
    countOfGenre.append(sum(Test_data['Genre'] == Genre_list[i]))

plt.figure(figsize = (10, 10))
plt.barh(Genre_list, countOfGenre)

for i in range(len(Genre_list)):
    plt.annotate(countOfGenre[i], xy = (countOfGenre[i], i), va = "center")
    
plt.ylabel("Genre")
plt.xlabel("Count per Genres")
plt.title("Test Data Count per Genres")
plt.show()
```
